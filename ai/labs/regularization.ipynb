# Regularization

```python
import numpy as np
import matplotlib.pyplot as plt

def polynomial(x, coefficients):
    # Starting with the first coefficient times x to the power of 0, ...
    return sum(c * x**i for i, c in enumerate(coefficients))

# Generate x values
x = np.linspace(-1.5,1.5, 400)

# Define coefficients for different degrees of polynomial
high_coefficients = np.array([35, -650, -50, 170, 20, 60, 175, -15, -105])

# Divide the high coefficients by 5 to get the low coefficients
low_coefficients = high_coefficients / 5

zero_coefficients = np.zeros(9)

# Calculate y values for all sets of coefficients
y_zero = polynomial(x, zero_coefficients)
y_low = polynomial(x, low_coefficients)
y_high = polynomial(x, high_coefficients)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(x, y_high, label='High Coefficients: f(x) = -650x - 50x^2 + 170x^3 + 20x^4 + 60x^5 + 175x^6 - 15x^7 - 105x^8 + 35')
plt.plot(x, y_low, label='Low Coefficients: f(x) = -130x - 10x^2 + 34x^3 + 4x^4 + 12x^5 + 35x^6 - 3x^7 - 21x^8 + 7')
plt.plot(x, y_zero, label='Zero Coefficients: f(x) = 35')
plt.title('Impact of Coefficient Values on Polynomial Smoothness')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
