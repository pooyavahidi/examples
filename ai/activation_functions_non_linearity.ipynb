# Activation Functions and Non-Linearity

## Activation Functions


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))


def relu(x):
    return np.maximum(0, x)


def linear_activation(x):
    return x
```

```python
def plot_activation(ax, x, y, title, axhline_value):
    ax.plot(x, y, linewidth=4)
    ax.set_xlabel("z")
    ax.set_ylabel("g(z)")
    ax.axvline(x=0, color="green", linestyle="--")
    ax.axhline(y=axhline_value, color="green", linestyle="--")
    ax.set_title(title)


# Create subplots
fig, axs = plt.subplots(1, 3, figsize=(18, 6))
x = np.linspace(-10, 10, 100)

plot_activation(
    axs[0], x, sigmoid(x), r"Sigmoid Function: $0 < g(z) < 1$", 0.5
)

plot_activation(axs[1], x, relu(x), r"ReLU Function: $g(z) = \max(0, z)$", 0)

plot_activation(
    axs[2],
    x,
    linear_activation(x),
    r"Linear Activation Function: $g(z) = z$",
    0,
)

plt.tight_layout()
plt.show()
```

## ReLU and Non-Linearity

```python
x = np.linspace(-1, 7, 400)

linear_piece_1 = relu(x)
linear_piece_2 = relu(2 * x - 4)
linear_piece_3 = relu(-0.8 * x + 3)
linear_piece_4 = relu(-2 * x + 10)
linear_piece_5 = relu(-1.5 * x + 9)

# Combine the pieces
y = (
    linear_piece_1
    + linear_piece_2
    + linear_piece_3
    + linear_piece_4
    + linear_piece_5
)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label="Piecewise Linear Function")
plt.axvline(
    x=0, color="green", linestyle="--", label="Transition Point at x=0"
)
plt.axvline(
    x=2, color="green", linestyle="--", label="Transition Point at x=2"
)
plt.axvline(
    x=4, color="green", linestyle="--", label="Transition Point at x=4"
)
plt.axvline(
    x=5, color="green", linestyle="--", label="Transition Point at x=5"
)
plt.axvline(
    x=6, color="green", linestyle="--", label="Transition Point at x=6"
)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Piecewise Linear Function (e.g. ReLU)")
plt.legend()
plt.show()
```

ReLU is defined as:
$$\text{ReLU}(x) = \max(0, x)$$

In the above example:<br>

- 1st linear piece: $\text{ReLU}(x)$

- 2nd linear piece: $\text{ReLU}(2x - 4)$

- 3rd linear piece: $\text{ReLU}(-0.8x + 3)$

- 4th linear piece: $\text{ReLU}(-2x + 10)$

- 5th linear piece: $\text{ReLU}(-1.5x + 9)$


So, now to find the overall function made of these linear pieces, we add them up:

$y= \text{ReLU}(x) + \text{ReLU}(2x - 4) + \text{ReLU}(-0.8x + 3) + \text{ReLU}(-2x + 10) + \text{ReLU}(-1.5x + 9)$


```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, label="Total Piecewise Linear Function", color="black")
plt.plot(x, linear_piece_1, label="First Linear Piece", linestyle="--")
plt.plot(x, linear_piece_2, label="Second Linear Piece", linestyle="--")
plt.plot(x, linear_piece_3, label="Third Linear Piece", linestyle="--")
plt.plot(x, linear_piece_4, label="Fourth Linear Piece", linestyle="--")
plt.plot(x, linear_piece_5, label="Fifth Linear Piece", linestyle="--")
plt.axvline(x=0, color="green", linestyle="--")
plt.axvline(x=2, color="green", linestyle="--")
plt.axvline(x=4, color="green", linestyle="--")
plt.axvline(x=5, color="green", linestyle="--")
plt.axvline(x=6, color="green", linestyle="--")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Piecewise Linear Function (e.g. ReLU)")
plt.legend()
plt.show()
```

The above shows each of the linear pieces and combined influence of each piece on the final function.
